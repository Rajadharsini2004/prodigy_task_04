```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/traffic accident dataset.csv')

print("Basic Information about the Dataset:")
print(data.info())

print("\nSummary Statistics for Categorical Columns:")
print(data.describe(include='object'))

categorical_column = 'Vehicle_movement'
print(f"\nFrequency Counts for {categorical_column}:")
print(data[categorical_column].value_counts())

plt.figure(figsize=(10, 6))
data[categorical_column].value_counts().plot(kind='bar', color='skyblue')
plt.title(f'Bar Plot for {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data[categorical_column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title(f'Pie Chart for {categorical_column}')
plt.show()
```

```python
categorical_column = 'Light_conditions'
print(f"\nFrequency Counts for {categorical_column}:")
print(data[categorical_column].value_counts())

plt.figure(figsize=(10, 6))
data[categorical_column].value_counts().plot(kind='bar', color='purple')
plt.title(f'Bar Plot for {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=data, x=categorical_column, palette='viridis')
plt.title(f'Countplot for {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Count')
plt.show()
```

```python
categorical_column = 'Type_of_collision'
print(f"\nFrequency Counts for {categorical_column}:")
print(data[categorical_column].value_counts())

plt.figure(figsize=(10, 6))
data[categorical_column].value_counts().plot(kind='bar', color='violet')
plt.title(f'Bar Plot for {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
counts = data[categorical_column].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4), colors=plt.cm.Paired.colors)
plt.title(f'Donut Chart for {categorical_column}')
plt.show()
```

```python

plt.figure(figsize=(8, 8))
data[categorical_column].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title(f'Pie Chart for {categorical_column}')
plt.show()
```

```python

another_categorical_column = 'Types_of_Junction'

plt.figure(figsize=(12, 6))
pd.crosstab(data[categorical_column], data[another_categorical_column]).plot(kind='bar', stacked=True, colormap='Paired')
plt.title(f'Stacked Bar Chart for {categorical_column} and {another_categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Count')
plt.show()
```
